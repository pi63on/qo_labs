```python
import numpy as np
import pandas as pd

import os
import glob
```

```python

# Specify the directory containing the files

# Use glob to list all files in the folder
files = glob.glob(os.path.join( './','*.txt'))  # * to get all files, or '*.ext' for specific file types

data = np.zeros(len(files), dtype=object)

for i,file in enumerate(files):
    with open(file, 'r') as f:
        content = f.read()
        print(f"Processing file: {file}")
        # You can now process the content as needed
        # Example: print the content
        print(content)
        data[i] = content
```

```python
ratios = np.zeros((len(data), 2))

for i in range(len(data)):
    print(len(data[i]))
    count0 = 0
    count1 = 1
    for number in data[i]:
        if int(number) == 0:
            count0 += 1
        else:
            count1 += 1
        
    ratios[i][0] = count0 / (count0+count1) 
    ratios[i][1] = count1 / (count0+count1) 

print(np.average(ratios, axis=0))
print(np.std(ratios, axis=0))
```
